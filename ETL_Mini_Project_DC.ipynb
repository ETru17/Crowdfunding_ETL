```python
# Import the dependency.
import pandas as pd
```

```python
#  Read in the dataset. 
crowdFunding_df = pd.read_excel("../Starter_Files/Resources/crowdfunding.xlsx")
crowdFunding_df 
```

```python
# Splitting the "category & sub-category" column into 2 separate columns
crowdFunding_df[['category', 'sub-category']] = crowdFunding_df['category & sub-category'].str.split('/', expand=True)

# Dropping the combined column and just keeping the separated columns
crowdFunding_df = crowdFunding_df.drop(columns=['category & sub-category'])
crowdFunding_df
```

```python
#grabbing unique categories
unique_categories = crowdFunding_df['category'].unique()

#creating category Data Frame
category_df = pd.DataFrame({
    'category_id': [f'cat{i+1}' for i in range(len(unique_categories))],
    'category': unique_categories
})
category_df
```

```python
#grabbing unique sub-categories
unique_subCategories = crowdFunding_df['sub-category'].unique()

#creating sub-category Data Frame
subCategory_df = pd.DataFrame({
    'subcategory_id': [f'cat{i+1}' for i in range(len(unique_subCategories))],
    'subcategory': unique_subCategories
})
subCategory_df
```
