```python
# Import the dependency.
import pandas as pd
```

```python
#  Read in the dataset. 
crowdFunding_df = pd.read_excel("../Starter_Files/Resources/crowdfunding.xlsx")
crowdFunding_df 
```

```python
# Splitting the "category & sub-category" column into 2 separate columns
crowdFunding_df[['category', 'sub-category']] = crowdFunding_df['category & sub-category'].str.split('/', expand=True)

# Dropping the combined column and just keeping the separated columns
crowdFunding_df = crowdFunding_df.drop(columns=['category & sub-category'])
crowdFunding_df
```

```python
#grabbing unique categories
unique_categories = crowdFunding_df['category'].unique()

#creating category Data Frame
category_df = pd.DataFrame({
    'category_id': [f'cat{i+1}' for i in range(len(unique_categories))],
    'category': unique_categories
})
category_df
```

```python
#Extracting the category DataFrame as "category.csv"
category_df.to_csv('category.csv', index=False)
```

```python
#grabbing unique sub-categories
unique_subCategories = crowdFunding_df['sub-category'].unique()

#creating sub-category Data Frame
subCategory_df = pd.DataFrame({
    'subcategory_id': [f'subcat{i+1}' for i in range(len(unique_subCategories))],
    'sub-category': unique_subCategories
})
subCategory_df
```

```python
#Extracting the sub-category DataFrame as "subcategory.csv"
subCategory_df.to_csv('subcategory.csv', index=False)
```

```python
crowdFunding_df
```

```python
# Merging category df to crowdfunding df
#sample merge used grocery_database = grocery_df_clean.merge(grocery_suppliers_df, on='upc12', how='right')
catcampaign_df = crowdFunding_df.merge(category_df, on='category', how='left')
catcampaign_df
```

```python
#merging to include both sub and category id columns
subcampaign_df = catcampaign_df.merge(subCategory_df, on='sub-category', how='left')
subcampaign_df
```

```python
# Creating the campaign Data Frame
campaign_df = pd.DataFrame({
    'cf_id': crowdFunding_df['cf_id'],
    'contact_id': crowdFunding_df['contact_id'],
    'company_name': crowdFunding_df['company_name'],
    'description': crowdFunding_df['blurb'],
    'goal': crowdFunding_df['goal'].astype(float),
    'pledged': crowdFunding_df['pledged'].astype(float),
    'outcome': crowdFunding_df['outcome'],
    'backers_count': crowdFunding_df['backers_count'],
    'country': crowdFunding_df['country'],
    'currency': crowdFunding_df['currency'],
    'launch_date': pd.to_datetime(crowdFunding_df['launched_at'], unit='s'),
    'end_date': pd.to_datetime(crowdFunding_df['deadline'], unit='s'),
    'category_id': subcampaign_df['category_id'],
    'subcategory_id': subcampaign_df['subcategory_id']
})
campaign_df
```

```python
#Extracting the campaign DataFrame as "campaign.csv"
campaign_df.to_csv('campaign.csv', index=False)
```
